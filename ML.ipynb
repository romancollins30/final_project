```python
import keras.preprocessing as kp
import os
import pandas as pd
import numpy as np
import PIL
```

```python
df = pd.DataFrame()
path = "Faces/500 Real Faces/PNG/"
filelist = os.listdir(path)
for x in filelist:
    if x.endswith(".png"):
        img = kp.image.load_img(path+x, target_size=(227, 227))
        y = kp.image.img_to_array(img)
        y = y.transpose(2,0,1).reshape(1,-1)
        df2 = pd.DataFrame(y)
        df = df.append(df2)
```

```python
df = df.reset_index()
df = df.drop(columns=['index'])
```

```python
lst = [1] * 500
df5 = pd.DataFrame(lst, columns=['real/fake'])
```

```python
main_df = pd.concat([df,df5], axis=1)
main_df.head()
```

```python
df = pd.DataFrame()
path = "images/"
filelist = os.listdir(path)
for x in filelist:
    if x.endswith(".png"):
        img = kp.image.load_img(path+x, target_size=(227, 227))
        y = kp.image.img_to_array(img)
        y = y.transpose(2,0,1).reshape(1,-1)
        df2 = pd.DataFrame(y)
        df = df.append(df2)
```

```python
df = df.reset_index()
df = df.drop(columns=['index'])
```

```python
lst = [0] * 100
df5 = pd.DataFrame(lst, columns=['real/fake'])
```

```python
main_df2 = pd.concat([df,df5], axis=1)
main_df2.head()
```

```python
final_df =main_df.append(main_df2)
final_df = final_df.reset_index()
final_df = final_df.drop(columns=['index'])
final_df
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
target = final_df["real/fake"]
data = final_df.drop("real/fake", axis=1)
feature_names = data.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
