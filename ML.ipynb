```python
import keras.preprocessing as kp
import os
import pandas as pd
import numpy as np
```

```python
#img = kp.image.load_img('linkedin_profile_pic.jpg', target_size=(227, 227))
#x = kp.image.img_to_array(img)
#x = x.transpose(2,0,1).reshape(1,-1)
df = pd.DataFrame()
path = "Faces/500 Real Faces/"
filelist = os.listdir(path)
for x in filelist:
    if x.endswith(".jpg"):
        img = kp.image.load_img(path+x, target_size=(227, 227))
        y = kp.image.img_to_array(img)
        y = y.transpose(2,0,1).reshape(1,-1)
        df2 = pd.DataFrame(y)
        df = df.append(df2)
```
